Clean the dataframe

```python
# Remove columns for IDs and links - not needed for the analysis

df.drop(['product_id', 'user_id', 'user_name', 'review_id', 'img_link', 'product_link' ], axis=1, inplace=True)
df.head(2)
```

Change datatype for some columns (from object to integer/float)

```python
# Notice that the some "supposed" numeric columns were set as object.
df.dtypes
```

```python
# Inspect discounted_price column

df.discounted_price.unique()
```

```python
# For discounted_price column, remove ‚Çπ and ","
# using str.replace() function


df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ', "")
df['discounted_price'] = df['discounted_price'].str.replace(',', "")

df.head()
```

```python
df.dtypes
```

```python
# Change discounted_price column data type to Float64
# using .astype() function

df['discounted_price'] = df['discounted_price'].astype('Float64')
```

```python
# Confirm discounted_price column

df.discounted_price.unique()
```

```python
df.head(2)
```

```python
# For actual_price column, remove ‚Çπ and ","
# using str.replace() function

df['actual_price'] = df['actual_price'].str.replace('‚Çπ', "")
df['actual_price'] = df['actual_price'].str.replace(',', "")

df.head(2)
```

```python
# Change actual_price column data type to Float64
# using .astype() function

df['actual_price'] = df['actual_price'].astype('Float64')
```

```python
# For discount_percentage column, remove %
# using str.replace() function

df['discount_percentage'] = df['discount_percentage'].str.replace('%',"")


df.head(2)
```

```python
# Change discount_percentage column data type to int64
# using .astype() function

df['discount_percentage'] = df['discount_percentage'].astype('int64')
```

```python
# Inspect column rating:
df.rating.unique()
```

```python
# Notice the unwanted character in the dataframe element

df[df.rating == '|']
```

```python
# Remove the row with the unwanted character in the rating column

indexRating = df[df['rating'] == '|'].index

df.drop(indexRating, inplace=True)
```

```python
# For rating column, change the data type to Float64
# using .astype() function

df['rating'] = df['rating'].astype('Float64')


df.dtypes
```

```python
# Inspect rating_count column

df.rating_count.unique()
```

```python
# For rating_count column, remove ","
# using str.replace() function

df['rating_count'] = df['rating_count'].str.replace(',', "")

df.head(2)
```

```python
# Change rating_count column data type to int
# using .astype() function

df['rating_count'] = df['rating_count'].astype('int')


df.head(2)
```

1. Check product type with its respective rating. What is highest and lowest rating of the product type mostly rated and reviewed?

```python
# product type breakdown of category
catType =[]
for categories in df['category'].values:
    catType += categories.split("|")
catType = list(dict.fromkeys(catType))
print(catType)
len(catType)
```

```python
# Count the frequency per catType

freq=[]
for cat in catType:
    counter = 0
    for temp in df['category']:
        if cat in temp.split('|'):
            counter += 1
    freq.append([cat, counter])
    
df_freq = pd.DataFrame(freq, columns= ["Category", "Frequency"])
df_freq.sort_values("Frequency", ascending = True, inplace = True)
```

```python
df_freq = df_freq.sort_values(by='Frequency', ascending=False)
```

```python
# Top 5 types of product types mostly rated and reviewed:

df_freq.head()
```

```python
# or you may view it graphically as:

df_freq.head(20).plot(kind='barh', x='Category', y='Frequency')
plt.xlabel('Frequency count')
```

3. What is price range of the products related to 'Home&Kitchen' with atleast 4.5 rating?
